```python
def DFS2(graph, v, LABEL): 
    '''
    Depth First Search(Traversal) for a graph

    :param v: vertex in graph
    '''
    
    graph[v][-1] = LABEL# Mark the current node as visited
    children = graph[v][1]
    if len(children)==1 :
        w,v2 = children[0]
        if graph[v2][-1]<0 and len(graph[v2][0])<=1: 
            DFS2(graph, v2, LABEL)

from plot import centroid
graph2 = graph.copy()
for v in graph2.keys():
    c = centroid(cntdict[v[0]][v[1]])
    graph2[v] += [c,-1]
    for iii in [0,1]:
        if len(graph2[v][iii])>=2:
            graph2[v][iii].sort(key = lambda x: -1*x[0])



uniquelabels = np.unique([g[2] for g in graph.values()])
for tID in uniquelabels:#[123]:
    for v in graph2.keys():
        if graph2[v][2]==tID and graph2[v][-1]<0:
            LABEL = graph2[v][3][1]
            DFS2(graph2, v, LABEL)


def tree(graph2, tID):
    '''
    for x,y,ls in lines:
        plt.plot(x,y,ls,color='grey')
    '''
    lines = []
    xz = {}
    for v in graph2.keys():
        if graph2[v][2]==tID:
            if graph2[v][-1] not in xz.keys():
                xz[graph2[v][-1]] = [v[0]]
            else:
                xz[graph2[v][-1]].append(v[0])
            for iii in [0,1]:
                if len(graph2[v][iii])>=2:
                    for w,v2 in graph2[v][iii]:
                        z0,z1 = v[0]*tRes,v2[0]*tRes
                        x0,x1 = graph2[v][-1]*0.1,graph2[v2][-1]*0.1
                        lines.append([[z0,z1],[x0,x1],0.3])
    xrange,zrange = [],[]
    for x,zlst in xz.items():
        x = x*0.1
        zmin,zmax = min(zlst)*tRes,max(zlst)*tRes
        if zmin==zmax:
            zmax = zmin+tRes/10
        xrange.append(x)
        zrange.append(zmin)
        zrange.append(zmax)
        lines.append([[zmin,zmax],[x,x],1])
    zmid,xmid = (min(zrange)+max(zrange))/2,(min(xrange)+max(xrange))/2
    return lines,zmid,xmid


FW,FH = 8,11
fig = plt.figure(figsize=(FW,FH),dpi=300)
                    

# (${\mu}m$)
# -------------------- simple --------------------
loc = {'L':0.5,'W':3*0.5,'T':0.5,'H':3*0.5}
axes = fig.subplots(ncols=3, nrows=3, sharex='none', sharey='none',
                    gridspec_kw={'left':loc['L']/FW, 'right':(loc['L']+loc['W'])/FW, 
                                'top':1-loc['T']/FH,'bottom':1-(loc['T']+loc['H'])/FH,
                                'wspace':0.1, 'hspace':0.2})

axes[0][0].set(ylabel='15 ${\mu}m$')    
sampleIDlst = [114,77,102]
info = [[255,256,257],[169,170,172],[254,255,256]]
for ii,pzlst in enumerate(info):
    tID = sampleIDlst[ii]
    #pzlst = np.arange(pz0,pz1)#list(trajs[tID].keys())[pz0:pz1]
    xrange,yrange=[],[]
    for i,pz in enumerate(pzlst):
        ax = axes[ii,i]
        immask = np.zeros(imshape[1:],dtype=np.uint8)
        for fpxy in trajps[tID][pz-1]:
            fpx,fpy = [np.squeeze(arr) for arr in np.split(np.array(fpxy),[1],axis=1)]
            #x,y = close_cnt(cnt)
            immask[fpx,fpy] = 1
            xrange+=list(fpy)
            yrange+=list(fpx)
            ax.imshow(immask,cmap='binary',vmax=2.5)
            #ax.fill(x,y,color='grey')
    xmean,ymean = (min(xrange)+max(xrange))/2,(min(yrange)+max(yrange))/2
    xypad = 75
    for ax in axes[ii]:
        ax.set(xlim=[xmean-xypad,xmean+xypad],ylim=[ymean+xypad,ymean-xypad],aspect=1)
        #ax.set(xticks=np.arange(xmean-xypad,xmean+xypad+1,50),xticklabels=np.arange(0,2*xypad/10,5).astype(int))
        #ax.set(yticks=np.arange(ymean-xypad,ymean+xypad+1,50),yticklabels=np.arange(0,2*xypad/10,5).astype(int))
        ax.set(xticks=[],yticks=[])
for ax in axes[0]:
    ax.set(xticklabels=[])
for ax in axes[1]:
    ax.set(xticklabels=[])
for ax in axes[:,1]:
    ax.set(yticklabels=[])
for ax in axes[:,2]:
    ax.set(yticklabels=[])        
# -------------------- simple tree --------------------
loc = {'L':2.55,'W':1,'T':0.5,'H':1.5}
axes = fig.subplots(ncols=1, nrows=3, sharex='none', sharey='none',
                    gridspec_kw={'left':loc['L']/FW, 'right':(loc['L']+loc['W'])/FW, 
                                'top':1-loc['T']/FH,'bottom':1-(loc['T']+loc['H'])/FH,
                                'wspace':0, 'hspace':0.2})
for i,tID in enumerate(sampleIDlst[:3]):
    lines,xmean,ymean = tree(graph2, tID)
    ax = axes[i]
    for x,y,alpha in lines:
        ax.plot(x,y,alpha=alpha,color='grey')
    xpad,ypad = 20,5
    ax.set(xlim=[xmean-xpad,xmean+xpad],ylim=[ymean+ypad,ymean-ypad])
    ax.set(xticks=np.arange(xmean-xpad,xmean+xpad+1,10),xticklabels=np.arange(0,2*xpad,10))
    #ax.set(yticks=np.arange(ymean-ypad,ymean+ypad+1,5),yticklabels=np.arange(0,2*ypad,5).astype(int))
    #ax.scatter(xmean-xpad+5,ymean+ypad-2,color=clst[i],s=50,edgecolors='grey')
    ax.set(yticks=[])
axes[1].set(ylabel='Position')
axes[2].set(xlabel='Time (s)')
axes[0].set(xticklabels=[])
axes[1].set(xticklabels=[])


# -------------------- medium --------------------
loc = {'L':0.5,'W':8*0.6,'T':2.65,'H':2*0.6}
axes = fig.subplots(ncols=8, nrows=2, sharex='all', sharey='all',
                    gridspec_kw={'left':loc['L']/FW, 'right':(loc['L']+loc['W'])/FW, 
                                'top':1-loc['T']/FH,'bottom':1-(loc['T']+loc['H'])/FH,
                                'wspace':0.1, 'hspace':0.2})

axes[0][0].set(ylabel='30 ${\mu}m$')
axes = axes.ravel()

tID = 81
pzlst = [170,171,172,174,176,178,179,181,
          183,185,188,194,198,199,202,205]
#list(trajs[tID].keys())[pz0:pz1]
xrange,yrange=[],[]
for i,pz in enumerate(pzlst):
    ax = axes[i]
    immask = np.zeros(imshape[1:],dtype=np.uint8)
    for fpxy in trajps[tID][pz]:
        fpx,fpy = [np.squeeze(arr) for arr in np.split(np.array(fpxy),[1],axis=1)]
        #x,y = close_cnt(cnt)
        immask[fpx,fpy] = 1
        xrange+=list(fpy)
        yrange+=list(fpx)
        ax.imshow(immask,cmap='binary',vmax=2.5)
        #ax.fill(x,y,color='grey')
xmean,ymean = (min(xrange)+max(xrange))/2,(min(yrange)+max(yrange))/2
xypad = 150
for ax in axes:
    ax.set(xlim=[xmean-xypad,xmean+xypad],ylim=[ymean+xypad,ymean-xypad],aspect=1)
    #ax.set(xticks=np.arange(xmean-xypad,xmean+xypad+1,100),xticklabels=np.arange(0,2*xypad/10,10).astype(int))
    #ax.set(yticks=np.arange(ymean-xypad,ymean+xypad+1,100),yticklabels=np.arange(0,2*xypad/10,10).astype(int))
    ax.set(xticks=[],yticks=[])

# -------------------- medium tree --------------------
loc = {'L':5.75,'W':1.5,'T':2.65,'H':0.8}
ax = fig.subplots(ncols=1, nrows=1, sharex='none', sharey='none',
                    gridspec_kw={'left':loc['L']/FW, 'right':(loc['L']+loc['W'])/FW, 
                                'top':1-loc['T']/FH,'bottom':1-(loc['T']+loc['H'])/FH,
                                'wspace':0, 'hspace':0})
lines,xmean,ymean = tree(graph2, tID)
for x,y,alpha in lines:
    ax.plot(x,y,alpha=alpha,color='grey')
xpad,ypad = 14.5,5
ax.set(xlim=[xmean-xpad,xmean+xpad],ylim=[ymean+ypad,ymean-ypad])
ax.set(xticks=np.arange(xmean-xpad,xmean+xpad+1,10),xticklabels=np.arange(0,2*xpad,10).astype(int))
#ax.set(yticks=np.arange(ymean-ypad,ymean+ypad+1,5),yticklabels=np.arange(0,2*ypad+1,5).astype(int))
#ax.scatter(xmean-xpad+1.5,ymean+ypad-1.5, color=clst[3],s=50,edgecolors='grey')
ax.set(yticks=[])
ax.set(xlabel='Time (s)', ylabel='Position')

# -------------------- medium travel --------------------
loc = {'L':0.5,'W':8*0.6,'T':4.3,'H':2*0.6}
axes = fig.subplots(ncols=8, nrows=2, sharex='all', sharey='all',
                    gridspec_kw={'left':loc['L']/FW, 'right':(loc['L']+loc['W'])/FW, 
                                'top':1-loc['T']/FH,'bottom':1-(loc['T']+loc['H'])/FH,
                                'wspace':0.1, 'hspace':0.2})

axes[0][0].set(ylabel='30 ${\mu}m$')
axes = axes.ravel()

tID = 157
pzlst = [346, 349, 352, 355, 358, 362, 366, 370, 
         373, 378,382,385, 388, 392, 400, 406]
#list(trajps[tID].keys())[::3]
#list(trajs[tID].keys())[pz0:pz1]
xrange,yrange=[],[]
for i,pz in enumerate(pzlst):
    ax = axes[i]
    immask = np.zeros(imshape[1:],dtype=np.uint8)
    for fpxy in trajps[tID][pz]:
        fpx,fpy = [np.squeeze(arr) for arr in np.split(np.array(fpxy),[1],axis=1)]
        #x,y = close_cnt(cnt)
        immask[fpx,fpy] = 1
        xrange+=list(fpy)
        yrange+=list(fpx)
        ax.imshow(immask,cmap='binary',vmax=2.5)
        #ax.fill(x,y,color='grey')
xmean,ymean = (min(xrange)+max(xrange))/2,(min(yrange)+max(yrange))/2
xypad = 150
for ax in axes:
    ax.set(xlim=[xmean-xypad,xmean+xypad],ylim=[ymean+xypad,ymean-xypad],aspect=1)
    #ax.set(xticks=np.arange(xmean-xypad,xmean+xypad+1,100),xticklabels=np.arange(0,2*xypad/10,10).astype(int))
    #ax.set(yticks=np.arange(ymean-xypad,ymean+xypad+1,100),yticklabels=np.arange(0,2*xypad/10,10).astype(int))
    ax.set(xticks=[],yticks=[])
# -------------------- medium tree --------------------
loc = {'L':5.75,'W':1.5,'T':4.3,'H':0.8}
ax = fig.subplots(ncols=1, nrows=1, sharex='none', sharey='none',
                    gridspec_kw={'left':loc['L']/FW, 'right':(loc['L']+loc['W'])/FW, 
                                'top':1-loc['T']/FH,'bottom':1-(loc['T']+loc['H'])/FH,
                                'wspace':0, 'hspace':0})
lines,xmean,ymean = tree(graph2, tID)
for x,y,alpha in lines:
    ax.plot(x,y,alpha=alpha,color='grey')
xpad,ypad = 22.5,12.5
ax.set(xlim=[xmean-xpad,xmean+xpad],ylim=[ymean+ypad,ymean-ypad])
ax.set(xticks=np.arange(xmean-xpad,xmean+xpad+1,10),xticklabels=np.arange(0,2*xpad,10).astype(int))
#ax.set(yticks=np.arange(ymean-ypad,ymean+ypad+1,5),yticklabels=np.arange(0,2*ypad+1,5).astype(int))
#ax.scatter(xmean-xpad+1.5,ymean+ypad-1.5, color=clst[3],s=50,edgecolors='grey')
ax.set(yticks=[])
ax.set(xlabel='Time (s)', ylabel='Position')


# -------------------- distribution --------------------
loc = {'L':0.8,'W':2.1,'T':5.9,'H':3}
axes = fig.subplots(ncols=1, nrows=2, sharex='col', sharey='none',
                    gridspec_kw={'left':loc['L']/FW, 'right':(loc['L']+loc['W'])/FW, 
                                'top':1-loc['T']/FH,'bottom':1-(loc['T']+loc['H'])/FH,
                                'wspace':0, 'hspace':0.3})
ax = axes[0]
ad = np.array(list(trajad.values()))
alst,dlst = ad[:,0],ad[:,1]#list of total area covered by traj
ax.scatter(alst,dlst,c='grey',edgecolors='none',alpha=0.5)

ax.set(ylabel='Duration (s)',ylim=[1,200])
ax.set_yscale("log")

ax = axes[1]
abins,asums = area
asums = asums/asums.sum()
ax.plot(abins,asums,':',color='grey')
ax.plot(abins,np.cumsum(asums),'-',color='grey')
ax.set(ylabel='Fraction',ylim=[0,1.05],yticks=[0,0.5,1])
import matplotlib
for ax in axes:
    ax.set_xscale("log")
    ax.set(xlim=[1,2000])
axes[1].set(xlabel='Area (${\mu}m^2$)')



fig.savefig(pjoin(pathlst['plots'],'spread4.pdf'), format='pdf', transparent=True, rasterized=True)
```
